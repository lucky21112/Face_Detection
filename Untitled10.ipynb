**Image Preprocessing**

```python
image=cv2.resize(image,(800,600))
```

Resizing image and converting image to gray image as color image has more pixels( 600x800x3) as compared to grey or black and white (600x800)

```python
image_grey=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(image_grey)
```

**Face Detection using HaarCascade**

```python
detector=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

Here we have used two parameter one is scale factor and minNeighbors to find the faces in the image we have adjusted the value based on the image.





```python
detect=detector.detectMultiScale(image_grey,scaleFactor=1.3, minNeighbors = 4)

```

Detect contain coordinates of face box where first column represent X coordinate , second column represent Y coordinate while third and fourth column represent size of face

```python
print(detect)
```

**Plotting rectangular box around faces**

```python
for (x,y,w,h) in detect:
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(image)
```
